```python
df_ABC = pd.read_excel('dataABCxyz.xlsx', header=0, usecols="A:G")
# display(df_ABC)
df_new = df_ABC.copy()

df_new = df_new.melt(id_vars= ['Товар'], value_vars=['Январь', "Февраль",	"Март",	"Апрель",	"Май",	"Июнь"], value_name='Sales', var_name='Month', ignore_index=False)
# display(df_ABC.head(2))
# # print(list(df_ABC.columns[1:len(df_ABC.columns)]))
df_new.head(2)
# df_new['Товар'].unique()
df_new
```

# ABC вариант цикла


```python
abc_per = []
good_list = pd.DataFrame({'good': df_new['Товар'].unique()})
for i in range (good_list.shape[0]):
    abc = sum(df_new['Sales'][df_new['Товар'] == good_list['good'][i]])/sum(df_new['Sales'])
    abc_per.append(round(abc*100,1))
good_list['abc_per'] = abc_per
good_list = good_list.sort_values(by = 'abc_per', ascending = False)
# display(good_list)

good_list_new = good_list.reset_index(drop = True)
good_list_new['abc_per_accum'] = good_list_new['abc_per'].cumsum() # Куммулята ошибается из-за разной сортировки долей
good_list_new['abc_group'] = good_list_new['abc_per_accum'].apply(lambda x: 'A' if x < 80 else ('B' if x < 95 else 'C'))
good_list_new
```

# Обратное развертывание таблицы

```python
stacked_df = df_new.pivot(index='Товар', columns='Month', values="Sales")

stacked_df = stacked_df.reset_index()
# # # # stacked_df.reset_index().info()
# # # # stacked_df.reset_index().unstack()
stacked_df['Total_sales'] = stacked_df['Апрель']+stacked_df['Июнь']+stacked_df['Май']+stacked_df['Февраль']+stacked_df['Январь']+stacked_df['Март']
stacked_df['Share'] = stacked_df['Total_sales']/stacked_df['Total_sales'].sum()
stacked_df['Rank'] = stacked_df['Total_sales'].rank(ascending=False)
# # # # stacked_df.index.rename = ['Num']
# # stacked_df.reset_index(drop=True).head()
stacked_df
```

```python
# stacked_df[stacked_df['Rank']>5]

# lambda x:stacked_df[stacked_df['Rank']>x][Share].sum()
# stacked_df.query('Rank>5')
# stacked_df.filter(like='S', axis=1)

stacked_df['Kummul'] = stacked_df['Rank'].apply(lambda x: stacked_df[stacked_df['Rank']<=x]['Share'].sum())
# stacked_df.query('Rank <= 7')
def ABC_assigment(x):
    if x <0.8: return 'A'
    elif x < 0.95: return 'B'
    else: return 'C'


stacked_df["ABC"] = stacked_df['Kummul'].apply(ABC_assigment)
# display(stacked_df.head(5))

df_ABC_second = df_new.groupby('Товар', as_index=False).agg(Total_sales = ('Sales', 'sum')) #.sort_values('Total_sales', ascending=False)
df_ABC_second['Share'] = round(df_ABC_second['Total_sales']/df_ABC_second['Total_sales'].sum(),3)
df_ABC_second['Rank'] = df_ABC_second['Total_sales'].rank(ascending=False)
df_ABC_second['Kummul'] = round(df_ABC_second['Rank'].apply(lambda x: df_ABC_second[df_ABC_second['Rank']<=x]['Share'].sum()),2)
df_ABC_second['GrABC'] = df_ABC_second['Kummul'].apply(ABC_assigment)
df_ABC_second
```

# XYZ анализ


```python
df_XYZ = df_new.groupby('Товар', as_index=False).agg(SKO = ('Sales', 'std'), Mean = ('Sales', 'mean'))
df_XYZ['VAR'] = df_XYZ['SKO']/df_XYZ['Mean']

# # df_XYZ
def XYZfn(x):
    if x < 0.1:
        return 'X'
    elif x < 0.25:
        return 'Y'
    else:
        return 'Z'
    
df_XYZ['XYZ'] = df_XYZ['VAR'].apply(XYZfn)
# # display(df_XYZ)
df_over = pd.merge(stacked_df[['Товар', 'ABC']], df_XYZ[['Товар', 'XYZ']], on='Товар')
df_over['Group'] = df_over['ABC'] + df_over['XYZ']
df_over.sort_values(by='Group', ascending=True)

```

```python
df_ABC_adid = df_adidas_1.groupby('Retailer', as_index=False).agg(sum_Operating_Profit = ('Operating Profit','sum'))
df_ABC_adid['Share'] = df_ABC_adid['sum_Operating_Profit']/df_ABC_adid['sum_Operating_Profit'].sum()
df_ABC_adid['Rank'] = df_ABC_adid['Share'].rank(ascending=False)
df_ABC_adid['Kummul'] = df_ABC_adid['Rank'].apply(lambda x: df_ABC_adid[df_ABC_adid['Rank']<=x]['Share'].sum())
df_ABC_adid['abc_group'] = df_ABC_adid['Kummul'].apply(lambda x: 'A' if x < 0.80 else ('B' if x < 0.95 else 'C'))
df_ABC_adid
```
